```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams.update({'font.size': 22})
```

```python
def load_stats(algName, nState):
    path = './results/'
    name = f"{algName}_riverSwim_{nState}.npy"
    stats = np.load(path+name)
    return stats
```

```python
def plot(ucfh, psrl, nState, N=10):
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    fig.suptitle(f'RiverSwim with {nState} states')
    rewards_u = np.convolve(ucfh[:, 0], np.ones(N)/N, mode='valid')
    rewards_p = np.convolve(psrl[:, 0], np.ones(N)/N, mode='valid')
    
    axes[0].plot(rewards_u, color='blue', alpha=0.5)
    axes[0].plot(rewards_p, color='red', alpha=0.5)
    axes[0].set_xlabel('# of episodes')
    axes[0].set_ylabel('Reward (per episode)')
    
    axes[1].plot(ucfh[N-1:, 1], rewards_u, color='blue', alpha=0.5, label='UCFH')
    axes[1].plot(psrl[N-1:, 1], rewards_p, color='red', alpha=0.5, label='PSRL')
    axes[1].set_xlabel('elapsed time (in ms)')
    axes[1].legend(bbox_to_anchor=(1.05, 1), loc='upper left')
```

```python
nState = 3
ucfh = load_stats('ucfh', nState)
psrl = load_stats('psrl', nState)
plot(ucfh, psrl, nState)
```

```python
nState = 6
ucfh = load_stats('ucfh', nState)
psrl = load_stats('psrl', nState)
plot(ucfh, psrl, nState)
```
